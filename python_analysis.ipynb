```python
#Q 5 & 6. Correlation 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from CSV
df = pd.read_csv('correlation.csv')

# Calculate the correlation matrix
correlation_matrix = df[['Avg_Price', 'Avg_Mileage', 'Avg_CC', 'Avg_max_power', 'Avg_Torque']].corr()
print(correlation_matrix)
# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Car Features')
plt.show()

```

```python
# Q7. The petrol distribution

# Load data from CSV
df = pd.read_csv('selling.csv')

# Filter data for Petrol vehicles
petrol_df = df[df['fuel'] == 'Petrol']

# Function to calculate IQR
def calculate_iqr(series):
    return series.quantile(0.75) - series.quantile(0.25)

# Compute descriptive statistics for Petrol vehicles
petrol_group = petrol_df['selling_price']
median_price = petrol_group.median()
iqr_price = calculate_iqr(petrol_group)
outliers = ((petrol_group < petrol_group.quantile(0.25) - 1.5 * iqr_price) |
            (petrol_group > petrol_group.quantile(0.75) + 1.5 * iqr_price)).sum()

# Create a summary DataFrame
summary = pd.DataFrame({
    'Median Selling Price': [median_price],
    'IQR': [iqr_price],
    'Number of Outliers': [outliers]
}, index=['Petrol'])

print(summary)

# Create a box plot for Petrol vehicles with a logarithmic scale on the y-axis
plt.figure(figsize=(6, 9))  # Adjusted width
sns.boxplot(y='selling_price', data=petrol_df, color='green')
plt.yscale('log')  # Apply logarithmic scale to y-axis
plt.title('Distribution of Selling Prices for Petrol Vehicles')
plt.ylabel('Selling Price (log scale)')
plt.show()

```
